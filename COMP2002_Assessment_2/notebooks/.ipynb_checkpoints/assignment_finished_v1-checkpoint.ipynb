```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

print("SECTION A: IMPORTS")
print("Pandas is needed for data handling, StandardScaler to scale features, and train_test_split to split the dataset")
print()

print("SECTION B: LOAD DATA")
df = pd.read_csv('glass_dataset.csv') 
print("Load the DF")
print("Dataset shape (rows, columns):", df.shape)
print("Column names:", list(df.columns))
print()

print("SECTION C: BASIC SUMMARY")
print("First 5 rows preview of data:")
print(df.head())
print()

print("SECTION D: DATA TYPES & MISSING VALUES")
df['Type'] = df['Type'].astype('category')
print("Convert 'Type' to a category for classification tasks.")
print("Data types:")
print(df.dtypes)
print("Missing value counts in each column:")
print(df.isnull().sum())
print()

print("SECTION E: DESCRIPTIVE STATISTICS")
print("Numeric distributions (min, mean, max and so on).")
print(df.describe())
print()

print("SECTION F: FEATURES (X) AND TARGET (y)")
X = df.drop(columns='Type')
y = df['Type']
print("Separate the input features from the target for modeling.")
print("Features shape:", X.shape)
print("Target shape:", y.shape)
print()

print("SECTION G: SCALING FEATURES")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Scaling ensures that features are on a comparable scale for many models.")
print("Sample of scaled features heres a preview of 5 rows:")
print(pd.DataFrame(X_scaled, columns=X.columns).head())
print()

print("SECTION H: TRAIN/TEST SPLIT")
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
print("Keep data aside for measure model performance objectively.")
print("X_train shape:", X_train.shape)
print("X_test  shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test  shape:", y_test.shape)
print()

print("Task 1.1")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("SECTION B: KNN CLASSIFICATION")
print("Train multiple kNN models with different numbers of neighbors (k) and find which k gives the best accuracy then evaluate the chosen model.")
best_k = None
best_acc = 0.0
k_values = range(1, 21)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    if acc > best_acc:
        best_acc = acc
        best_k = k

print(f"Best k found: {best_k} with accuracy: {best_acc:.3f}")

knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X_train, y_train)
y_pred_knn = knn_best.predict(X_test)

print("KNN Accuracy on test set:", accuracy_score(y_test, y_pred_knn))
print("KNN Confusion Matrix:")
cm_knn = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues')
plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
print("KNN Classification Report:")
print(classification_report(y_test, y_pred_knn, zero_division=0))
print()

print("SECTION C: DEMO PREDICTION WITH KNN")
print("Predict on a new input to demonstrate how the trained KNN model makes a prediction.")
new_input = np.array([[1.517, 13.20, 3.50, 1.20, 72.7, 0.60, 8.2, 0.0, 0.0]])
knn_pred_new = knn_best.predict(new_input)
print("New input prediction (KNN):", knn_pred_new)
print()

print("SECTION D: SVM CLASSIFICATION")
print("Investigate the many different kernel shapes: linear, polynomial (degree 2 and 3), and RBF, then choose the best performer.")
kernels = [
    ('linear', None),
    ('poly', 2),
    ('poly', 3),
    ('rbf', None)
]
best_svm_acc = 0.0
best_svm_config = None
svm_models = {}

for kernel, degree in kernels:
    if kernel == 'poly':
        svm_clf = SVC(kernel=kernel, degree=degree, C=1.0)
    else:
        svm_clf = SVC(kernel=kernel, C=1.0)
    svm_clf.fit(X_train, y_train)
    y_pred_svm = svm_clf.predict(X_test)
    acc_svm = accuracy_score(y_test, y_pred_svm)
    svm_models[(kernel, degree)] = svm_clf
    if acc_svm > best_svm_acc:
        best_svm_acc = acc_svm
        best_svm_config = (kernel, degree)

print(f"Best SVM kernel configuration: {best_svm_config}, Accuracy: {best_svm_acc:.3f}")
best_kernel, best_degree = best_svm_config
best_svm = svm_models[(best_kernel, best_degree)]

y_pred_svm_best = best_svm.predict(X_test)
print("SVM Accuracy on test set:", accuracy_score(y_test, y_pred_svm_best))
print("SVM Confusion Matrix:")
cm_svm = confusion_matrix(y_test, y_pred_svm_best)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Purples')
plt.title("SVM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm_best, zero_division=0))
print()

print("SECTION E: DEMO PREDICTION WITH SVM")
svm_pred_new = best_svm.predict(new_input)
print("New input prediction (SVM):", svm_pred_new)
print()
print("Task 1.2")
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score

print("SECTION A: CROSS-VALIDATION FOR KNN")
print("We will assess how well our best kNN model (best_k) performs using 5-fold cross-validation.")
knn_scores = cross_val_score(knn_best, X, y, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores (kNN):", knn_scores)
print("Mean accuracy (kNN):", knn_scores.mean())
print("Standard deviation (kNN):", knn_scores.std())
print()

print("SECTION B: CROSS-VALIDATION FOR SVM")
print("We will assess how well our best SVM model (best_kernel/best_degree) performs with 5-fold cross-validation.")
svm_scores = cross_val_score(best_svm, X, y, cv=5, scoring='accuracy')
print("Cross-validation accuracy scores (SVM):", svm_scores)
print("Mean accuracy (SVM):", svm_scores.mean())
print("Standard deviation (SVM):", svm_scores.std())
print()

print("SECTION C: COMBINED BOX PLOT OF KNN AND SVM CV SCORES")
print("We plot the distribution of kNN and SVM accuracies across the 5 folds in a single figure.")
scores_list = [knn_scores, svm_scores]
plt.boxplot(scores_list, tick_labels=['kNN', 'SVM'])
plt.title("kNN vs. SVM Cross-Validation Accuracy (5-fold)")
plt.ylabel("Accuracy")
plt.show()
print()

print("SECTION E: SUMMARY & ANALYSIS")
print("After running 5-fold cross-validation on our kNN and SVM models, we see the results above:")
print("- kNN's mean CV accuracy is around", round(knn_scores.mean(), 3), 
      "with a standard deviation of", round(knn_scores.std(), 3))
print("- SVM's mean CV accuracy is around", round(svm_scores.mean(), 3), 
      "with a standard deviation of", round(svm_scores.std(), 3))
print()
print("We observe that kNN currently performs better on average than SVM, ")
print("though SVM might benefit from further hyperparameter tuning.")
print("Cross-validation gives us multiple accuracy scores (one per fold), ")
print("which is more robust than a single train/test split.")
print("Looking at the box plot, we can compare the distribution of accuracies for both models, ")
print("noting any outliers or differences in variance.")
print()

print("Task 1.3")
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("SECTION A: IMPORTS")
print("We import numpy for random number generation and array handling,")
print("and matplotlib for creating the plots.")
print()

print("SECTION B: DEFINE FITNESS FUNCTIONS")
def schaffer_n2(x, y):
    return 0.5 + (np.sin((x**2) - (y**2))**2 - 0.5) / (1 + 0.001*(x**2 + y**2))**2

def modified_ackley(x, y):
    return -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) \
           - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) \
           + 20

def easom(x, y):
    return -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))

print("We define three separate functions (Schaffer N.2, Modified Ackley, and Easom) ")
print("that each take x and y as inputs and return a single fitness value.")
print()

print("SECTION C: GENERATE RANDOM SOLUTIONS")
num_points = 500
print(f"We generate {num_points} random (x, y) points uniformly in [-5, 5] for each function.")

s_schaffer = np.random.uniform(-5, 5, (num_points, 2))
s_ackley   = np.random.uniform(-5, 5, (num_points, 2))
s_easom    = np.random.uniform(-5, 5, (num_points, 2))

f_schaffer = [schaffer_n2(x, y) for x, y in s_schaffer]
f_ackley   = [modified_ackley(x, y) for x, y in s_ackley]
f_easom    = [easom(x, y) for x, y in s_easom]
print("Fitness values calculated for each random point.")
print()

print("SECTION D: PLOT RESULTS")
print("We plot each problem in a separate subplot, coloring points by their fitness values.")
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Schaffer
axs[0].scatter(s_schaffer[:,0], s_schaffer[:,1], c=f_schaffer, cmap='viridis')
axs[0].set_title("Schaffer N.2")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")

# Ackley
sc_ackley = axs[1].scatter(s_ackley[:,0], s_ackley[:,1], c=f_ackley, cmap='viridis')
axs[1].set_title("Modified Ackley")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")

# Easom
sc_easom = axs[2].scatter(s_easom[:,0], s_easom[:,1], c=f_easom, cmap='viridis')
axs[2].set_title("Easom")
axs[2].set_xlabel("x")
axs[2].set_ylabel("y")

fig.colorbar(axs[0].collections[0], ax=axs[0], label="Fitness Value")
fig.colorbar(sc_ackley, ax=axs[1], label="Fitness Value")
fig.colorbar(sc_easom, ax=axs[2], label="Fitness Value")

plt.tight_layout()
plt.show()
print()

print("Task 2.1")
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("SECTION A: INITIAL SETUP")
print("Define the objective functions, set parameters, and initialise the population.")

def schaffer_n2(x, y):
    return 0.5 + (np.sin((x**2 - y**2))**2 - 0.5) / (1 + 0.001 * (x**2 + y**2))**2

def modified_ackley(x, y):
    return -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) \
           - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) \
           + 20

def easom(x, y):
    return -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))

fitness_function = easom # change this if wanted to another func 
pop_size = 100                   
num_generations = 50           
num_offspring = 100             
mutation_sigma = 0.1            
bounds = [-5, 5]              

parents = np.random.uniform(bounds[0], bounds[1], (pop_size, 2))
print(f"Initial population of {pop_size} individuals created within bounds {bounds}.")
print()

print("SECTION B: UNIFORM CROSSOVER FUNCTION")
def uniform_crossover(parents, num_offspring):
    offspring = []
    for _ in range(num_offspring):
        p1, p2 = parents[np.random.choice(len(parents), 2, replace=False)]
        mask = np.random.rand(2) < 0.5
        child = np.where(mask, p1, p2)
        offspring.append(child)
    return np.array(offspring)

print("This function randomly chooses each gene (x or y) from one of the two parents.")
print()

print("SECTION C: GAUSSIAN MUTATION FUNCTION")
def gaussian_mutation(offspring, sigma):
    mutation = np.random.normal(0, sigma, offspring.shape)
    mutated = offspring + mutation
    return np.clip(mutated, bounds[0], bounds[1])

print("Gaussian mutation completed")
print()

print("SECTION D: SELECTION FUNCTION")
def select_next_generation(parents, offspring, fitness_func, num_survivors):
    combined = np.vstack((parents, offspring))
    fitness = np.array([fitness_func(x, y) for x, y in combined])
    best_indices = np.argsort(fitness)[:num_survivors]  # lower fitness = better
    return combined[best_indices], fitness[best_indices]

print("This function combines parent and child populations and selects the best individuals.")
print()

print("SECTION E: RUN EVOLUTIONARY ALGORITHM")
best_fitness_per_gen = []
avg_fitness_per_gen = []

for gen in range(num_generations):
    offspring = uniform_crossover(parents, num_offspring)
    mutated_offspring = gaussian_mutation(offspring, mutation_sigma)
    parents, fitness = select_next_generation(parents, mutated_offspring, fitness_function, pop_size)

    best_fitness = fitness.min()
    avg_fitness = fitness.mean()

    best_fitness_per_gen.append(best_fitness)
    avg_fitness_per_gen.append(avg_fitness)

    print(f"Generation {gen+1}: Best = {best_fitness:.6f}, Avg = {avg_fitness:.6f}")

print()

print("SECTION F: PLOT RESULTS")
plt.plot(best_fitness_per_gen)
plt.title(f"Best Fitness per Generation ({fitness_function.__name__})")
plt.xlabel("Generation")
plt.ylabel("Best Fitness")
plt.grid(True)
plt.show()

print("Task 2.2")
```

```python
print("SECTION G: VISUALISATION OF RESULTS")
print("Plot the average values over all generations to observe the optimiser's performance.")

plt.figure(figsize=(10, 5))
plt.plot(avg_fitness_per_gen, label="Average Fitness", color='orange')
plt.title(f"Fitness Progression over Generations ({fitness_function.__name__})")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.legend()
plt.grid(True)
plt.show()

print("Task 2.3")
```
